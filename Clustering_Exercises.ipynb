**Clustering Exercises**

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
```

```python
filename = '/content/cust_seg.csv'
customers = pd.read_csv(filename)
cust_df = customers[['Age', 'Income']]
cust_df.head()
```

```python
# Instantiate Standard Scaler
scaler = StandardScaler()
# Fit & transform data.
scaled_df = scaler.fit_transform(cust_df)

```

```python
from sklearn.cluster import AgglomerativeClustering
# Use Hierarchial Clustering
hc = AgglomerativeClustering(n_clusters = 2)
hc.fit(scaled_df)

```

```python
# Save the cluster lables to the dataframe 
cust_df['cluster'] = hc.labels_
# Visualize the clusters
plt.scatter(cust_df['Age'], cust_df['Income'], c = cust_df['cluster'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Clusters of Customers');



```

```python
#Use DB Scan
# Instantiate Standard Scaler
scaler = StandardScaler()
# Fit & transform data.
scaled_df = scaler.fit_transform(cust_df)
```

```python
from sklearn.cluster import DBSCAN
# Instantiate & fit clustering - this is done in one step for DBSCAN
dbs = DBSCAN(eps = 0.5, min_samples = 5).fit(scaled_df)

```

```python
# Save the cluster lables to the dataframe 
cust_df['cluster'] = dbs.labels_
# Visualize the clusters
plt.scatter(cust_df['Age'], cust_df['Income'], c = cust_df['cluster'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Clusters of Customers');
```

```python
#Use KMeans Clustering
#Import
import seaborn as sns
#Scatter Plot
sns.scatterplot(x=customers['Age'], y=customers['Income'])
```

```python
# Instantiate KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(scaled_df)
```

```python
cust_df['cluster'] = kmeans.labels_

# Visualize the clusters
plt.scatter(cust_df['Age'], cust_df['Income'], c = cust_df['cluster'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Clusters of Customers');
```

**Which clustering algorithm is best for this data and why?**

KMeans clustering because it separates the data points of Age and Income to other columns.

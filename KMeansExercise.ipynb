```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

filename = '/content/cust_seg.csv'
customers = pd.read_csv(filename)
cust_df = customers[['Age', 'Income']]
cust_df.head()
```

```python
# Instantiate Standard Scaler
scaler = StandardScaler()
# Fit & transform data.
scaled_df = scaler.fit_transform(cust_df)

```

```python
# Visualize Dendogram
plt.figure(figsize = (15, 5))
sch.dendrogram(sch.linkage(scaled_df, method = 'ward'))
plt.xlabel('Data Points');
```

```python
#Import Data
import seaborn as sns
#Scatter Plot
sns.scatterplot(x=customers['Age'], y=customers['Income'])
```

```python
# Instantiate KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(scaled_df)
```

```python
cust_df['cluster'] = kmeans.labels_

# Visualize the clusters
plt.scatter(cust_df['Age'], cust_df['Income'], c = cust_df['cluster'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Clusters of Customers');
```
